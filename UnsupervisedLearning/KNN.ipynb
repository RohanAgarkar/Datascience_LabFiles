```python
import pandas
import numpy
import matplotlib.pylab as plt
%matplotlib inline
import seaborn

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
data = make_blobs(n_samples=500, n_features=2, centers=4, cluster_std=1.5, random_state=50)
```

```python
data
```

```python
plt.scatter(data[0][:, 0], data[0][:,1],c=data[1],cmap="rainbow")
```

```python
model = KMeans(n_clusters=4,n_init=10, max_iter=1000)
model.fit(data[0], data[1])
```

```python
model.cluster_centers_
```

```python
model.labels_
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c=model.labels_, cmap="rainbow")
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:, 1], marker="^", color="black")
```

```python
silhouette_score(data[0], model.labels_)
```
