```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.linspace(0, 1, 11)
y = x ** 2
```

```python
plt.plot(x, y)
```

```python
# subplots
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(x, y, 'r--') # more color options later
plt.title("chart 1") # to show title
plt.subplot(1,2,2)
plt.plot(y,x, 'g*-')
plt.title("chart 2") # to show title
```

```python
# Matplotlib ObjectOrientedMethod

fig = plt.figure()

# add set of axes to figure
axes = fig.add_axes([0.1, 0.1,0.5,0.5]) # left, bottom, width, height(range 0 to 1)\

# plot on theat set of axis
axes.plot(x,y, 'b')
axes.set_xlabel('Set X Label')
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
```

```python
# Matplotlib ObjectOrientedMethod

fig = plt.figure()

# add set of axes to figure
axes = fig.add_axes([0.1, 0.1,0.8,0.8]) # main axes
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) # inset axes

# plot on theat set of axis
axes.plot(x,y, 'b--')
axes.set_xlabel('Set X Label')
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')

# Insert figure Axes 2

axes2.plot(y, x, 'r-*')
axes2.set_xlabel("X_label_axes2")
axes2.set_ylabel("y_label_axes2")
axes2.set_title("Axes 2 title")
```

```python
# subplots
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y, 'r')
axes[0].set_xlabel('Set X Label')
axes[0].set_ylabel('Set y Label')
axes[0].set_title('Set Title')

axes[1].plot(y, x, 'ro')
axes[1].set_xlabel("X_label_axes2")
axes[1].set_ylabel("y_label_axes2")
axes[1].set_title("Axes 2 title")

plt.tight_layout()
```
