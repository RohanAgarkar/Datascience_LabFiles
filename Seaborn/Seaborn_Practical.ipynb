## Seaborn Practical

```python
import seaborn as sns
import pandas as pd
df = sns.load_dataset('tips')
```

```python
sns.displot(df.total_bill, kde=False, bins=15)
```

```python
sns.jointplot(x=df.total_bill, y='tip', data=df, kind='scatter')
```

```python
sns.jointplot(x=df.total_bill, y='tip', data=df, kind='reg')
```

```python
sns.pairplot(df, palette='viridis')
```

```python
sns.rugplot(df['total_bill'])
```

```python
sns.heatmap(df.corr(numeric_only=True), annot = True)
```

```python
sns.barplot(x='day', y='total_bill', data=df)
```

```python
sns.boxplot(x='day', y='total_bill', data=df, palette='flare')
```

```python
sns.boxplot(data=df, palette='flare', orient='h')
```

```python
sns.swarmplot(x='day', y='total_bill', data=df, hue='smoker', palette='bone')
```

```python
sns.stripplot(x='day', y='total_bill', data=df, hue='smoker', palette='bone')
```

## Working on Titanic Dataset

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
sns.displot(x=df.sex, y=df.fare)
```

```python
sns.jointplot(x="class", y='fare', data=df, palette="flare")
```

```python
sns.jointplot(x='class', y='fare', data=df)
```

```python
sns.pairplot(df, palette='flare')
```

```python
sns.rugplot(x='fare', y='class', data=df)
```

```python
sns.barplot(x='pclass', y='fare', data=df)
```

```python
sns.swarmplot(data=df, x='pclass', y='fare')
```

```python
sns.stripplot(x='sex', y='age', data=df)
```
